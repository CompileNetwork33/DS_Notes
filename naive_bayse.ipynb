Naive Bayes Algorithm

Machine Learning ka ek bahut famous and fast classification algorithm hai.

Ye mainly:

text classification
spam detection
sentiment analysis

me use hota hai.

🧠 Core Idea

Naive Bayes probability pe based hota hai.

Ye predict karta hai:

“Given these features, kis class ki probability sabse zyada hai?”

🔥 Why Name = “Naive” Bayes?
🔷 Bayes

Because it uses:

Bayes Theorem

🔷 Naive

Because ye assume karta hai:

saare features independent hain

Real world me mostly ye assumption wrong hoti hai 😄
But surprisingly algorithm fir bhi bahut acha kaam karta hai.

Suppose email me words hain:

🔷 Working Steps
Step 1

Training data se probabilities learn karta hai.

Step 2

New input ke liye:

P(Class∣Features)

calculate karta hai.

Step 3

Highest probability wali class choose karta hai.

🔥 Types of Naive Bayes
1. Gaussian Naive Bayes

Continuous numerical data ke liye.

Example:

height
weight
salary

Assumes:
👉 data Gaussian distribution follow karta hai.

2. Multinomial Naive Bayes

Count/frequency based data ke liye.

Mostly:

NLP
text classification

Example:

word counts
3. Bernoulli Naive Bayes

Binary features ke liye.

Example:

word present or absent
yes/no
📊 Example

Suppose:

Outlook	Play Cricket
Sunny	No
Rainy	Yes

Algorithm probabilities calculate karega:

Sunny → No
Rainy → Yes
🔥 Advantages ✅
Very fast
Works well on small data
Excellent for text data
Simple implementation
Less training time
🔥 Disadvantages ❌
Independence assumption unrealistic hoti hai
Complex relationships samajh nahi pata
Zero probability issue
🔥 Zero Probability Problem

Agar training data me koi feature kabhi appear hi nahi hua:
👉 probability = 0

To solve:
✅ Laplace Smoothing use hota hai.
