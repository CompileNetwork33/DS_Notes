#Feature scaling (Standralization)

Scaling ka matlab hota hai features (columns) ke values ko ek similar range me lana.

Example:

Age = 18 to 70

Salary = 10,000 to 10,00,000

Yahan Salary ka range bahut bada hai. Agar hum directly model train karein, to model Salary ko zyada importance de dega, jo galat ho sakta hai.

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(df[["AGE"]])
```

```python
ss.transform(df[["AGE"]])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(ss.transform(df[["AGE"]]))
plt.show()
```

```python
plt.subplot(1,2,1)
plt.title("Before Scaling")
sns.histplot(ss.transform(df[["AGE"]]))

plt.subplot(1,2,2)
plt.title("After Scaling")
sns.histplot(df[["AGE"]])
plt.show()

```
