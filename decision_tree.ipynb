## KNN

KNN ka logic bahut human-like hai 😄

“Nearby cheezein usually similar hoti hain.”

Example:
Agar tumhare aas-paas mostly doctors hain,
to chances hain tum medical environment se related ho 👀

Machine learning me bhi same idea.

Meaning of “K”

K = number of nearest neighbors.

Example:

K = 3
K = 5

Algorithm nearby K points dekhta hai.

🔷 How KNN Works

Suppose new point classify karna hai.

Step 1

K choose karo.

Example:

K = 3
Step 2

New point aur sab training points ke beech distance calculate karo.

Usually:
👉 Euclidean Distance

use hota hai.
d=\sqrt{(x_2-x_1)^2+(y_2-y_1)^2}


Nearest K neighbors select karo.

Step 4

Majority vote lo.

Example:
Nearest neighbors:

Dog
Dog
Cat

✅ Final prediction = Dog


🔥 Choosing K Value
Small K

Example:

K = 1

👉 Very sensitive to noise

Can overfit.

Large K

👉 Smoother predictions

But underfitting ho sakti ha

#### For scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
sc.fit(x)
x = pd.DataFrame(sc.transform(x),columns=x.columns)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=4)
```

```python
knn.fit(x_train, y_train)
```

```python
knn.score(x_test,y_test)
```

```python
knn.score(x_train,y_train)
```

```python
for i in range(1,6):
    knn1 = KNeighborsClassifier(n_neighbors=1)
    knn1.fit(x_train, y_train)
    print(i, knn1.score(x_train, y_train),knn1.score(x_test, y_test))
```

```python
from mlxtend import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=knn1)
```
