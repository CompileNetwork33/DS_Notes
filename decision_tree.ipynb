## KNN

KNN ka logic bahut human-like hai 😄

“Nearby cheezein usually similar hoti hain.”

Example:
Agar tumhare aas-paas mostly doctors hain,
to chances hain tum medical environment se related ho 👀

Machine learning me bhi same idea.

Meaning of “K”

K = number of nearest neighbors.

Example:

K = 3
K = 5

Algorithm nearby K points dekhta hai.

🔷 How KNN Works

Suppose new point classify karna hai.

Step 1

K choose karo.

Example:

K = 3
Step 2

New point aur sab training points ke beech distance calculate karo.

Usually:
👉 Euclidean Distance

use hota hai.
d=\sqrt{(x_2-x_1)^2+(y_2-y_1)^2}


Nearest K neighbors select karo.

Step 4

Majority vote lo.

Example:
Nearest neighbors:

Dog
Dog
Cat

✅ Final prediction = Dog


🔥 Choosing K Value
Small K

Example:

K = 1

👉 Very sensitive to noise

Can overfit.

Large K

👉 Smoother predictions

But underfitting ho sakti ha
