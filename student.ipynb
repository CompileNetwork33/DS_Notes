One Hot Encoding kya hota hai?

Categorical data (Male, Female, City, etc.) ko numeric binary columns (0/1) me convert karna = One Hot Encoding.

Example:
Gender
Male
Female

After One Hot Encoding:

Gender_Male	Gender_Female
1	0
0	1
🔹 Kyun zaroori hai?

ML models numbers samajhte hain, text nahi ❌
Isliye categorical → numeric banana padta hai ✔️

🔹 sklearn me OneHotEncoder
from sklearn.preprocessing import OneHotEncoder

🔹 Column names ke saath DataFrame

🔹 Dummy Trap kya hota hai?

Agar sab categories ke columns rakh liye →
columns aapas me dependent ho jaate hain (multicollinearity) ❌

Example (Gender):

Male + Female = 1 (always)

Isse model confuse hota hai.

✅ Solution:

ohe = OneHotEncoder(sparse=False, drop='first')


Ek column drop ho jayega → dummy trap avoid ✔️

📌 PART 2: DUMMY VARIABLES (Concept)

Dummy Variables bhi same kaam karti hain:
Categorical → 0/1 columns

Pandas me:

pd.get_dummies(df['Gender'])


With drop first:

pd.get_dummies(df['Gender'], drop_first=True)

📌 PART 3: INTERVIEW QUESTIONS & ANSWERS 🔥
❓ Q1. What is One Hot Encoding?

✅ Answer:
One Hot Encoding is a technique to convert categorical variables into binary numeric columns so that machine learning models can process non-numeric data.

❓ Q2. Why do we use One Hot Encoding?

✅ Answer:
Because machine learning algorithms work with numerical data and cannot directly handle categorical text values.

❓ Q3. What are dummy variables?

✅ Answer:
Dummy variables are binary (0/1) columns created from categorical features to represent the presence or absence of a category.

❓ Q4. What is the dummy variable trap?

🔥 Very important interview question

✅ Answer:
Dummy variable trap is a situation where dummy variables are highly correlated with each other, causing multicollinearity in regression models.

❓ Q5. How do you avoid dummy variable trap?

✅ Answer:
By dropping one category from each encoded feature using drop='first' in OneHotEncoder or drop_first=True in pandas get_dummies.

❓ Q6. Difference between OneHotEncoder and LabelEncoder?

✅ Answer:

OneHotEncoder	LabelEncoder
Creates multiple binary columns	Assigns numbers (0,1,2…)
No ordinal meaning	Creates false order ❌
Best for nominal data	Only for target labels
❓ Q7. When should we NOT use One Hot Encoding?

✅ Answer:
When a categorical feature has very high cardinality (too many unique values), because it creates too many columns and increases model complexity.

❓ Q8. What problem occurs with high number of encoded columns?

✅ Answer:
It increases dimensionality, memory usage, training time and may cause overfitting.

❓ Q9. How does OneHotEncoder handle unseen categories?

✅ Answer:
By default it throws an error. We can handle this using:

OneHotEncoder(handle_unknown='ignore')

🧠 PLACEMENT TIP (very useful for you 💯)

In projects & interviews, best practice:

ohe = OneHotEncoder(
    sparse=False,
    drop='first',
    handle_unknown='ignore'
)
