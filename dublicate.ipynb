#Handling  of dublicate data

```python
import pandas as pd
```

```python
data = {"Name": ["Alice", "Bob", "Charlie", "Alice","Sita","Veeru","krishyu","Nancy","Vaishnavi","Aman","Manas"],"Eng":[99,66,44,99,94,47,90,98,56,54,88],"Maths":[89,99,98,89,55,89,87,90,67,54,79],"Gender":[1,0,0,1,0,1,1,1,0,+1,-0]}
```

```python
df = pd.DataFrame(data)
df
```

```python
df["duplicated"] = df.duplicated()
df
```

```python
df.loc[[0,3]]
```

```python
df.drop_duplicates()#dl krta 
df.drop_duplicates(inplace=True)#original data me dl kr deeta hai
df
```

### Changing of data types and handling of duplicate data

```python
df.info()
```

```python
df["Gender"].value_counts()#ginti karta hai kitne values kitni bar hai
```

```python
df["Gender"].fillna(df["Gender"].mode()[0])
```

```python
df["Gender"] = df["Gender"].astype("bool")#.astype data types change krta hai
```

```python
df.info()
```
