Data Cleaning in Machine Learning
1️⃣ Handling Missing Values

Sometimes datasets have NaN / null values.

***Methods***

1. Remove rows

df.dropna()

2. Fill with mean

df["Age"].fillna(df["Age"].mean())

3. Fill with median

df["Age"].fillna(df["Age"].median())

4. Fill with mode (for categorical data)

df["City"].fillna(df["City"].mode()[0])

 ## Missing values and how can find it
 In Machine Learning / Data Analysis, a missing value means data that is not stored or is unavailable in the dataset. Instead of an actual value, the dataset may contain NaN, NULL, or blank cells.

What is a Missing Value?

A missing value is when a feature has no data for some observations.

Example dataset:

Name	Age	Salary
Ravi	25	50000
Aman	NaN	60000
Neha	30	NaN

Here:

Age for Aman is missing

Salary for Neha is missing

In Python (Pandas), missing values are usually shown as NaN.

***Why Missing Values Occur***

Common reasons:

Data not collected

User skipped a form field

Data corruption

Errors during data entry

Data merging issues

Feature Selection Techniques in Machine Learning

1️⃣ Filter Methods

Isme features ko model train kiye bina statistical methods se select kiya jata hai.

Common Techniques

Correlation

Information Gain

Variance Threshold

Example

Agar kisi feature ki variance bahut kam hai, to uska data almost same hoga → us feature ka model me koi contribution nahi.

✔ Fast
✔ Simple
❌ Feature interaction consider nahi karta

Wrapper Methods:-

Is method me model ko train karke different feature combinations test kiye jate hain.

Techniques

Forward Selection

Backward Elimination

Recursive Feature Elimination (RFE)

**Forward Selection**

Start → 0 features

Ek-ek feature add karte hain

Jo model accuracy improve kare usko keep karte hain

Example

Start → none
Step1 → add Age
Step2 → add Salary
Step3 → add Experience

**Backward Elimination**

Start → all features

Least important feature remove karte hain

Example

Start → Age, Salary, Experience, City

Remove → City (least important)

***Recursive Feature Elimination (RFE)***

Model train hota hai

Least important features remove hote hain

Process repeat hota hai

3️⃣ Embedded Methods

Examples
